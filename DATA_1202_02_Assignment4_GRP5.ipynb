```python
#getting top 1000 rows
top_1000= youtube_df.iloc[:1000]
top_1000
```

```python
#Checking unique channeltype
top_1000['channeltype'].unique()
```

```python
#function to calculate the distribution of channeltype
def channeltype_distribution(df,grpbycol,coltocount):
    count_col= df.groupby(grpbycol)[coltocount].count()
    return count_col.sort_values(ascending=False)
```

```python
channeltype_count=channeltype_distribution(top_1000,'channeltype','name')
channeltype_count
```

```python
#loading top 1000 into new csv file
top_1000.to_csv('top1000YOUTUBE.csv')
```

```python
#load top 1000 to MySQL Database
#Create engine and connect
engine = create_engine('dialect+driver://username:password@host:port/database')
#create engine
conn = engine.connect()#connection stirng

#Import into MYSQL table
top_1000.to_sql('youtube_top1000', engine, index=False)
```
